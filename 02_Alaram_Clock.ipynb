```python
from datetime import datetime
import pygame
import time

# Ask user to enter time in 12-hour format with AM/PM
ala_time = input("Enter the time for alarm in HH:MM:SSAM/PM (e.g., 07:30:00PM): ")

print("Setting up alarm....")

while True:
    now = datetime.now().strftime("%I:%M:%S%p")  # current time in same format
    if now == ala_time.upper():
        print("Reminder!! Press ENTER to stop the alarm.")
        
        # Initialize pygame mixer
        pygame.mixer.init()
        pygame.mixer.music.load(r"C:\Users\kondu\OneDrive\Desktop\Python Projects\audio.mp3")
        pygame.mixer.music.play(-1)  # loop until stopped

        # Wait for user input to stop alarm
        input()  
        pygame.mixer.music.stop()
        print("Alarm stopped ✅")
        break

    time.sleep(1)
```
